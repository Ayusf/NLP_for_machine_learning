```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')  

lemmatizer = WordNetLemmatizer()
corpus = []

for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])  
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

### gpt example

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = ["I love pizza and pizza is tasty"]

# Without binary (default) → counts
cv_count = CountVectorizer(ngram_range=(1, 1))  # unigrams
X_count = cv_count.fit_transform(corpus).toarray()

print("Without binary (Count):")
print(cv_count.get_feature_names_out())
print(X_count)

# With binary=True → just presence
cv_binary = CountVectorizer(binary=True, ngram_range=(1, 1))
X_binary = cv_binary.fit_transform(corpus).toarray()

print("\nWith binary=True (Presence):")
print(cv_binary.get_feature_names_out())
print(X_binary)
```
