```python
import pandas as pd
message = pd.read_csv('SMSSpamCollection', sep='\t', names=['label','message'] )
```

```python
message
```

```python
import nltk
import re
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []

for msg in message['message']:
    # Keep only letters (remove numbers, punctuation), and convert to lowercase
    review = re.sub('[^a-zA-Z]', ' ', msg).lower()
    
    # Split into words
    words = review.split()
    
    # Remove stopwords and apply stemming
    cleaned_words = [ps.stem(word) for word in words if word not in stopwords.words('english')]
    
    # Join words back into a single string
    cleaned_review = ' '.join(cleaned_words)
    
    # Add the cleaned message to the corpus
    corpus.append(cleaned_review)
```

```python
corpus
```

```python
y=pd.get_dummies(message['label'])
y=y.iloc[:,0].values # .values convert into the array
```

```python
y
```

```python
## traintest split 
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(corpus, y, test_size=0.20)
print(xtrain,ytrain)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
xtrain = cv.fit_transform(xtrain).toarray()
xtest = cv.transform(xtest).toarray()
```
